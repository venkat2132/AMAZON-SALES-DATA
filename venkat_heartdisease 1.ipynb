```python
#Attribute Information:

#age
#sex
#chest pain type (4 values)
#resting blood pressure
#serum cholestoral in mg/dl
#fasting blood sugar > 120 mg/dl
#resting electrocardiographic results (values 0,1,2)
#maximum heart rate achieved
#exercise induced angina
#oldpeak = ST depression induced by exercise relative to rest
#the slope of the peak exercise ST segment
#number of major vessels (0-3) colored by flourosopy
#thal: 0 = normal; 1 = fixed defect; 2 = reversable defect
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(data_d.corr(),annot=True)
plt.show()
```

```python
sns.countplot(data=data_d,x='target')
plt.title('Heart-Disease rates')
plt.xticks([0,1],['No-disease','Disease'])
plt.show()
```

```python
sns.countplot(data=data_d,x='sex',hue='target')
plt.title('heart disease cause by gender')
plt.xticks([0,1],['female','male'])
plt.show()
```

```python
sns.distplot(data_d['age'],bins=20)
plt.title('heart disease cause by age')
plt.show()
```

```python
# Typical Angina (TA), 
# Atypical Angina (ATA), 
# Non-Anginal Pain (NAP)
# Asymptomatic (ASY)
```

```python
sns.countplot(data_d['cp'])
plt.title('chest pain types')
plt.xticks([0,1,2,3],['TA','ATA','NAP','ASY'])
plt.show()
```

```python
data_d['trestbps'].hist()
plt.title('Resting blood pressure in distribution')
plt.show()
# people who are having blood pressure between 120-130
```

```python
sns.countplot(data=data_d,x='fbs',hue='target')
plt.title('Fasting blood sugar')
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```
