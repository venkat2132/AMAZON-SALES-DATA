```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
data
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
data['Month'] = data['Order Date'].dt.month
data['Year']  = data['Order Date'].dt.year
data.head()
```

```python
sns.countplot(data=data,x='Sales Channel')
plt.title('Sales channel')
plt.show()
```

```python
sns.countplot(data = data, x = 'Month')
plt.title('Month wise oders')
plt.show()
```

```python
sns.countplot(data = data, x = 'Year')
plt.title('Year wise oders')
plt.show()
```

```python
data.groupby(data['Order Date'].dt.year)['Total Revenue'].sum().plot(kind='bar')
plt.title('Total revenue year-wise')
plt.show()
```

```python
data.groupby(data['Order Date'].dt.month)['Total Revenue'].sum().plot(kind='bar')
plt.title('Total revenue month-wise')
plt.show()
```

```python
plt.figure(figsize=(16,8))
data.groupby([data['Order Date'].dt.year, data['Month']])['Total Revenue'].sum().unstack().plot(kind='bar')
plt.title('Total revenue Year & month-wise')
plt.show()
```

```python
sns.countplot(data = data, y = 'Region')
plt.title('Region wise oders')
plt.show()
```

```python
sns.countplot(data = data, y = 'Item Type')
plt.title('item type sales')
plt.show()
```
